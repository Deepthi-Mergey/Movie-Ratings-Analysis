```python
# python is not a vectorize programming language
# Building dashboards (dashboard - combination of charts)

sns.set_style('darkgrid')
f, axes = plt.subplots (2,2, figsize = (15,15))

k1 = sns.kdeplot(df, x = 'BudgetMillions', y = 'AudienceRating', ax = axes[0,0])
k2 = sns.kdeplot(df, x = 'BudgetMillions', y = 'CriticRating', ax = axes[0,1])

k1.set(xlim = (-20,160))
k2.set(xlim = (-20,160))

z = sns.violinplot(df[df.Genre == 'Drama'], x = 'Year', y = 'CriticRating', ax = axes[1,0])

k4 = sns.kdeplot(df, x = 'CriticRating', y = 'AudienceRating', shade = True, shade_lowest = False, cmap ='Reds', ax = axes[1,1])

k4b = sns.kdeplot(df, x = 'CriticRating', y = 'AudienceRating', cmap = 'Reds', ax = axes[1,1])

plt.show()
```

```python
# How we can style the dashboard  using different color map

sns.set_style('dark',{'axes.facecolor':'black'})
f, axes = plt.subplots (2,2, figsize = (15,15))

#plot [0,0]
k1 = sns.kdeplot(df, x = 'BudgetMillions', y = 'AudienceRating', \
                 shade = True, shade_lowest = True, cmap = 'inferno', \
                 ax = axes[0,0])
k1b = sns.kdeplot(df, x = 'BudgetMillions', y = 'AudienceRating', \
                 cmap = 'cool', ax = axes[0,0])

#plot [0,1]
k2 = sns.kdeplot(df, x = 'BudgetMillions', y = 'CriticRating',\
                 shade = True, shade_lowest = True, cmap ='inferno',\
                 ax = axes[0,1])
k2b = sns.kdeplot(df, x = 'BudgetMillions', y = 'CriticRating',\
                  cmap = 'cool', ax = axes[0,1])

#plot[1,0]
z = sns.violinplot(df[df.Genre == 'Drama'], \
                   x = 'Year', y = 'CriticRating', ax = axes[1,0])

#plot[1,1]
k4 = sns.kdeplot(df, x = 'CriticRating', y = 'AudienceRating', \
                 shade = True, shade_lowest = False, cmap ='Blues_r', \
                 ax = axes[1,1])

k4b = sns.kdeplot(df, x = 'CriticRating', y = 'AudienceRating', \
                  cmap = 'gist_gray_r', ax = axes[1,1])


k1.set(xlim = (-20,160))
k2.set(xlim = (-20,160))

plt.show()
```
